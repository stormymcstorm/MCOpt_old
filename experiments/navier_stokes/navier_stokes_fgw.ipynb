```python
import sys

import matplotlib.pyplot as plt
import numpy as np

from mcopt.pipeline import Pipeline
from mcopt import MorseGraph, ot

sys.path.append('../')
from util import (
  draw_graphs, 
  run_fgw,
)
```

```python
cmap = 'cool'

dist = 'geo'
hist = 'degree'

alpha = 0.5
```

```python
pipeline = Pipeline('../../pipeline.config.json')

navier_stokes = pipeline.graph('navier_stokes')
```

```python
src = navier_stokes.frames[0]

dests = navier_stokes.frames.copy()
dests.pop(0)

len(dests)
```

# Initial Graphs

```python
fig = draw_graphs(
  src, dests, 
  width=3, 
  height=3, 
  cmap='cool', 
  src_title='Navier Stokes 0', 
  dest_title_fmt='Navier Stokes {i}'
)
```

# fGW Transport

```python
fgw_couplings = run_fgw(
  src, dests, alpha=alpha, hist=hist, dist=dist, random_G0=False
)
```

```python
fig = draw_graphs(
  src, dests, 
  width=3, 
  height=3, 
  cmap='cool', 
  src_title='Navier Stokes 0', 
  dest_title_fmt='Navier Stokes {i}',
  couplings=fgw_couplings
)

fig.suptitle(f'fGW Transport (\u03B1 = {alpha})', fontsize=80)
```
