```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

import numpy as np

from mcopt.pipeline import Pipeline
from mcopt import MorseGraph
```

```python
pipeline = Pipeline('../../pipeline.config.json')

sinusoidal_complex = pipeline.complex('sinusoidal').frames[0]
sinusoidal_noisy_complex = pipeline.complex('sinusoidal').frames[1]
```

```python
sinusoidal = sinusoidal_complex.to_graph()
sinusoidal_noisy = sinusoidal_noisy_complex.to_graph()
```

```python
cmap = ListedColormap('black')
```

```python
fig, ax = plt.subplots()

sinusoidal.draw(ax, cmap=cmap)
```

```python
fig, ax = plt.subplots()

sinusoidal_noisy.draw(ax, cmap=cmap)
```

```python
sinusoidal_sampled = sinusoidal.sample(rate=10, mode='step')

fig, ax = plt.subplots()

sinusoidal_sampled.draw(ax, cmap=cmap)
```

```python
sinusoidal_noisy_sampled = sinusoidal_noisy.sample(rate=10, mode='step')

fig, ax = plt.subplots()

sinusoidal_noisy_sampled.draw(ax, cmap=cmap)
```
